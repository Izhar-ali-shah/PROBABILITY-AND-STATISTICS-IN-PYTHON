```python
y = [6,7,2,4,5]
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(y)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(y)
plt.title("First example plot")
plt.xlable("x vals (s)")
plt.ylable("y vals (m)")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(y)
plt.title("First example plot")
plt.xlabel("x vals (s)")
plt.ylabel("y vals (m)")
plt.grid()
plt.show()
```
